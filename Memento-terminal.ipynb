# Memento sur l'utilisation du terminal

Ouvrir un terminal avec le raccourcis clavier sous linux.
```bash
    ctrl + alt + t
```
    

## Exécuter une commande

Par exemple afficher le répertoire courant

```shell
> pwd
/home/user
```

A cette commande on peut lui ajouter des paramètres. Par exemple le paramètre qui permet d'afficher la documentation liée à la commande est le paramètre --help

```shell
> pwd --help
pwd: pwd [-LP]
    Affiche le nom du répertoire de travail courant.
    
    Options :
      -L	affiche la valeur de $PWD s'il nomme le répertoire de travail
    		courant
      -P	affiche le répertoire physique, sans aucun lien symbolique
    
    Par défaut, « pwd » se comporte comme si « -L » était spécifié.
    
    Code de retour :
    Renvoie 0 à moins qu'une option non valable ne soit donnée ou que le
    répertoire courant ne peut pas être lu.
```

### Lister le contenu d'un dossier

Pour une énumération des fichiers/dossiers présent dans le dossier en cours.
```shell
> ls 
```

Pour un présentation en mode liste des fichiers/dossiers présent dans le dossier en cours.
```shell
> ls -l
```

Pour un présentation en mode liste des fichiers/dossiers "cachés" présent dans le dossier en cours.
```shell
> ls -la
```

Pour un présentation en mode liste des fichiers/dossiers présent dans le dossier passé en paramètre.
```shell
> ls mon-dossier/le-sous-dossier/ -l
```

### Changer de répertoire

Pour changer de répertoire. Ici on souhaite entrer dans le www contenu dans le dossier yyy qui est lui même contenu dans le dossier mon-dossier.
```shell
> cd mon-dossier/yyy/www
```

Pour remonter dans le répertoire parent direct. Ici on sera dans le dossier yyy
```shell
> cd ..
```

Pour revenir directement dans le dossier personnel de l'utilisteur, on utilise cd sans paramètre.
```shell
> cd
```

### Utilisation de l'historique de commande

Il suffit d'appuyer sur les touches <b>flêches du haut</b> ou du <b>bas</b> pour faire défiler les commandes déjà saisies. 
Pour effectuer une recherche dans l'historique on peut utiliser également le raccourcis <b>ctrl + r</b>.
Cela va permettre la saisi de la commande que l'on recherche.

### Autocomplétion

Pour autocompléter une commande, un chemin d'un répertoire, on utilise la touche "<b>tabulation</b>".
Pour avoir la liste des possibilités de l'autocomplétion il faut appuyer deux fois consécutive sur la touche tabulation.

### Se déplacer dans le prompt

Pour aller directement en début de la ligne de commande en cours de saisi on utilise le raccourcis <b>ctrl + a</b> et <b>ctrl + e</b> pour aller à la fin de la ligne de commande


### Créer un répertoire

Pour créer un répertoire dans le dossier courant on utilise la commande <b>mkdir</b>
```shell
> mkdir nouveauDossier
```

Si on souhaire utiliser les espaces dans le nom du dossier il faut entourer le nom du dossier par des guillemets. Sans ces guillemets la commande mkdir va comprende que l'on souhaite créer deux dossiers "nouveau" et "dossier"
```shell
> mkdir "nouveau dossier"
```

Si on souhaite créer directement une arborescence il faut ajouter le paramètre <b>p</b>. Cela va créer le dossier "nouveauDossier" dans le tata qui est dans le dossier titi qui est lui même dans le dossier toto. Le paramètre <b>p</b> va créer automatiquement les dossiers parents si ceux-ci n'existent pas.
```shell
> mkdir -p toto/titi/tata/nouveauDossier
```

### Créer un fichier

Pour créer un fichier dans le répertoire où l'on se trouve on utilise la commande <b>touch</b>
```shell
> touch fichier1.txt
```

Pour créer un fichier dans un dossier particulier.
```shell
> touch dossier/sous-dossier/fichier1.txt
```

### Déplacer un fichier / dossier

Pour déplacer un fichier dans un autre répertoire.
```shell
> mv fichier1.txt dossier/fichier1.txt
```

Pour déplacer un dossier dans un autre répertoire.
```shell
> mv dossier1/ dossier/sous-dossier/
```

Pour déplacer plusieurs fichiers qui possèdent un pattern commun dans un autre dossier. Ici tous les fichiers commençant par "fich" vont être déplacer dans le répertoire dossier/sous-dossier
```shell
> mv fich* dossier/sous-dossier
```

### Renommer un fichier / dossier

Si le dossier / fichier n'existe pas alors il peut être renommé via la commande <b>mv</b>.<br />
Ici le fichier1.txt va être renommé en clients.txt
```shell
> mv fichier1.txt clients.txt
```

### Copier un fichier

Pour copier un fichier on utilise la commande <b>cp</b>.<br />Ici on a copié le contenu du fichier1.txt dans un nouveau fichier produits.txt
```shell
> cp fichier1.txt produits.txt
```

### Copier un dossier

Pour copier un dossier avec tous les sous dossier qu'il contient on utilise la commande <b>cp</b> mais avec le paramètre -r.

```shell
> cp -r dossierACopier/ destination/
```

### Supprimer un fichier

Pour supprimer un fichier on utilise la commande <b>rm</b>.
```shell
> rm fichier1.txt
```

### Supprimer un dossier

Pour supprimer un dossier on utilise la commande <b>rm</b> avec le paramètre -r.
```shell
> rm -r dossierASupprimer
```

### Afficher le manuel d'une commande

On utilise la commande <b>man</b> suivi de la commande
Pour créer un fichier dans un dossier particulier.
```shell
> man ls
```
### Afficher le contenu d'un fichier

On utilise la commande <b>cat</b> pour afficher le contenu d'un fichier dans le terminal.
Pour créer un fichier dans un dossier particulier.
```shell
> cat dossier/sous-dossier/fichier1.txt
```

Pour afficher le contenu avec une "navigation". Pour sortir on appuie sur la touche "q".
```shell
> less dossier/sous-dossier/fichier1.txt
```

Pour afficher des dernières lignes d'un fichier en "temps réel" on utilise la commande <b>tail</b> avec le paramètre "-f"
```shell
> tail -f dossier/sous-dossier/fichier1.txt
```

### Rediriger le retour d'une commande

Pour rediriger le retour d'une commande dans un fichier on utilise la commande suivi d'un chevron puis le fichier de destination.
```shell
> ls -l dossier/ > log.txt
```

### Rechercher une information dans un fichier

Pour effectuer une recherche dans le contenu de fichier on utilise la commande <b>grep</b>. Ici on recherche le mot "Bonjour" dans le fichier monfichier.txt
```shell
> grep "Bonjour" monfichier.txt
```

Pour effectuer une recherche dans le contenu de tous les fichiers d'un dossier on utilise la commande <b>grep</b> avec le paramètre "-r". Ici on recherche le mot "Bonjour" dans tous les fichiers du dossier "dossier"
```shell
> grep -r "Bonjour" dossier/
```
